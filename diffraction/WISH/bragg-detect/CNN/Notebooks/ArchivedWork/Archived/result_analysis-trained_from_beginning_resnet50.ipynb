### Calculate False positive rate

```python
def calculate_confusion_matrix_per_bin(bin_data, real_bboxes, predicted_bboxes):
    real_peaks = []
    predicted_peaks = []
    for box in real_bboxes:
        x = tc.round((box[0]+box[2])/2)
        y = tc.round((box[1]+box[3])/2)
        real_peaks.append((x,y))

    for box in predicted_bboxes:
        x = tc.round((box[0]+box[2])/2)
        y = tc.round((box[1]+box[3])/2)
        predicted_peaks.append((x,y))

    print(f"Real peaks {real_peaks}")
    print(f"Pred peaks {predicted_peaks}")

    true_positives = 0
    false_positives = 0
    false_negatives = 0
    true_negatives = 0
    
    for peak in predicted_peaks:
        if peak in real_peaks:
            true_positives += 1
        else:
            false_positives += 1

    for peak in real_peaks:
        if peak not in predicted_peaks:
            false_negatives += 1

    N = bin_data.shape[0] * bin_data.shape[1]
    true_negatives = N - true_positives - false_positives - false_negatives
    return true_positives, false_positives, false_negatives, true_negatives
```

```python
r_model.eval()
probe = 0
for img, target in data_loader:
    imgs = list(img)
    targets = list(target)
    with tc.no_grad():
        for img, target in zip(imgs, targets):
            prediction = r_model([img.to(device)])[0]
            print(f"Original predictions={prediction}")
            nms_prediction = apply_nms(prediction, iou_thresh=0.001)
            print(f"NMS predictions={nms_prediction}")
            # plot_bin_detections(img[0,:,:], target['boxes'], nms_prediction, 0.50)
            TP, FP, FN, TN = calculate_confusion_matrix_per_bin(img[0,:,:], target['boxes'], nms_prediction['boxes'])
            print(TP, FP, FN, TN)
            probe += 1

    if probe >= 1:
        break
```
