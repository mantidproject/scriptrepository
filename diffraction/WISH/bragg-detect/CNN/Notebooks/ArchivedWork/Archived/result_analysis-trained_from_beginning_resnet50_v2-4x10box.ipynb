```python
import torch as tc
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import pickle
import utils
import os
from wish_utils import *
from WishDataSet import WishDataSet, get_transform
```

```python
annotations_dir = "/home/wj1132075/Desktop/CNN_Model_Data/PeakDataInFullRange/Annotations/"
```

```python
model_path = '/mnt/ceph/home/wj1132075/Desktop/CNN_Model_Data/fasterrcnn_resnet50_fpn_v2_trained_from_beginning_4x10box.pt'
```

```python
device = tc.device('cpu')
device
```

```python
dataset_test = WishDataSet(annotations_dir, transforms=get_transform(train=False))
data_loader = tc.utils.data.DataLoader(dataset_test, batch_size=8, shuffle=False, num_workers=2, collate_fn=utils.collate_fn)
```

```python
r_model = get_object_detection_model_v2(2)
r_model.load_state_dict(tc.load(model_path, map_location=device))
r_model = r_model.to(device)
```

```python
r_model.eval()
probe = 0
for img, target in data_loader:
    imgs = list(img)
    targets = list(target)
    with tc.no_grad():
        for img, target in zip(imgs, targets):
            prediction = r_model([img.to(device)])[0]
            print(f"Original predictions={prediction}")
            nms_prediction = apply_nms(prediction, iou_thresh=0.001)
            print(f"Predictions after NMS={len(nms_prediction['boxes'])}, {nms_prediction}")
            plot_bin_detections(img[0,:,:], target['boxes'], nms_prediction, 0.50)
            probe += 1

    if probe >= 50:
        break
```

### Calculate False positive rate

```python
from tqdm.notebook import tqdm
r_model.eval()
probe = 0

tot_TP = 0
tot_FP = 0
tot_FN = 0
tot_TN = 0
for img, target in tqdm(data_loader):
    imgs = list(img)
    targets = list(target)
    with tc.no_grad():
        for img, target in zip(imgs, targets):
            prediction = r_model([img.to(device)])[0]
            # print(f"Original predictions={prediction}")
            nms_prediction = apply_nms(prediction, iou_thresh=0.001)
            # print(f"NMS predictions={nms_prediction}")
            
            TP, FP, FN, TN = calculate_confusion_matrix_per_bin(img[0,:,:], target['boxes'], nms_prediction['boxes'], x_tol=2, y_tol=2)
            tot_TP += TP
            tot_FP += FP
            tot_FN += FN
            tot_TN += TN
            # print(TP, FP, FN, TN)
            # probe += 1

    # if probe >= 1:
    #     break
```

```python
tot_TP, tot_FP, tot_FN, tot_TN
```

```python
FPR = tot_FP/(tot_FP+tot_TN)*100
print(f"False positive rate={FPR}%")
```

```python
TPR = tot_TP/(tot_TP+tot_FN)*100
print(f"True positive rate={TPR}%")
```

```python
precision = tot_TP/(tot_TP+tot_FP)*100
print(f"precision={precision}%")
```

```python
recall = tot_TP/(tot_TP+tot_FN)*100
print(f"Recall {recall}%")
```
